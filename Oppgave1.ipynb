```python
import pandas as pb
import numpy as np

# If you want to use the csv from this git set
url = "https://raw.githubusercontent.com/umaimehm/Intro_to_AI_2021/main/Lab1/stud.csv"
# You can also download the csv and set

df = pb.read_csv(url, sep=',')
df.head()
df = df.dropna()

df.isna().sum()
df = df.replace(r'^\s*$', np.nan, regex=True)

df['FinalGrade'] = df['FinalGrade'].astype(str).astype(float)

df = df[df.FinalGrade <= 100]
df.info()
```

```python
for i in range(50):
    if df['FinalGrade'][i] > 90:
       print("SD")
```
